## Bland-Altman-Plots

e.g. https://cran.r-project.org/web/packages/BlandAltmanLeh/vignettes/Intro.html<br>


land-Altman plots are a well established method to check the agreement of different measurement methods or the retest-reliability of a single measurement method. (Apparently, Bland-Altman plots are otherwise known as Tukey’s Mean Difference Plot.) They do not come included in R but can easily be produced using R (find an overview of alternatives for this package at the end of this document). The BlandAltmanLeh package tries to make Bland-Altman plots even more accessible. It also features confidence intervals and the combination of Bland-Altman-/Tukey Mean Difference Plot and sunflower plot.

This package is based on the 1986 publication of JM Bland and DG Altman. It is available for free download on http://www-users.york.ac.uk/~mb55/meas/ba.htm Written for medical doctors, this article is very well readable for the non-statistician and thus highly recommended.

````r
y = data$age_at_scan
y_hat = data$y_preds

plot(y, y_hat, main="Scatter plot")
abline(0,1)
````

This is a useful plot but may often not be optimal. As the data will mostly be placed on a diagonal line, most of the space in the diagram is wasted. Deviations from identity are displayed on a diagonal line which may not be what our brain recognises best. Finally, the choice of which measurement is placed on the horizontal and which on the vertical axis is often arbitrary. Bland and Altman (and apparently Tukey before them) propose a different approach, where the x axis is the mean of two measurements each and the y axis is the difference between them.

````r
plot((y+y_hat)/2, y-y_hat, main="Mean-Difference-Plot")
````

````r
library(BlandAltmanLeh)
bland.altman.plot(y, y_hat, main="This is a Bland Altman Plot", xlab="Means", ylab="Differences")
````

### ggplot2

````r
library(ggplot2)
pl <- bland.altman.plot(y, y_hat, graph.sys = "ggplot2")
print(pl)
````

There are good reasons to prefer each of the graphics systems. Mostly it is a matter of taste.

About 18 out of 589 data points in the figure above falls out of the 95% confidence interval depicted by the upper and lower line. That’s just what one would expect in case of normal distribution.

````r
hist(y-y_hat)
````

**Confidence Intervals**

Of course, these lines have an error margin and Bland and Altman 1986 describe how to compute confidence intervals for the lines. These can also be calculated and printed with the BlandAltmanLeh package:

````r
pl <- bland.altman.plot(y, y_hat, graph.sys="ggplot2", conf.int=.95)
print(pl)
````

````r
# or in base-graphics:
bland.altman.plot(y, y_hat, conf.int=.95, pch=19)
````

````r
data$gender[1:10]
````

````r
data$sex<-recode(data$gender, 'F'=1, 'M'=2)
````

````r
ba.stats <- bland.altman.stats(y, y_hat)

plot(ba.stats$means, ba.stats$diffs, col=data$sex, 
     sub=paste("critical difference is", round(ba.stats$critical.diff,2)),
     main="By gender", ylim=c(-15,15), pch=18-data$sex)
abline(h = ba.stats$lines, lty=c(2,3,2), col=c("lightblue","blue","lightblue"), 
       lwd=c(3,2,3))
legend(x = "topright", legend = c("male","female"), fill = 1:2)
````

There is a lot of worthwhile additions available on CRAN. One could to point to the ggExtra package, which can be used to display marginal distributions at ggplot2-plots like:

````r
library(ggExtra)
print(ggMarginal(bland.altman.plot(y, y_hat, graph.sys = "ggplot2"),
           type = "histogram", size=4))
````
